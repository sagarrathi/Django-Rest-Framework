% floats: percentage of values that are float
% int: percentage of values that are whole numbers
% string: percentage of values that are strings
% unique string: number of unique string values / total number
% unique integers: number of unique integer values / total number
mean numerical value (non numerical values considered 0 for this)
std deviation of numerical values


```python

class Colu:
    def __init__(self,df, col, sorp):
        
        self.col=col
        self.data=np.array( list(df[col]) )
        self.sorp=sorp
        
        # Identify
        self.prediction="unknown"
        self.dtype="unknown"
        self.n=0
        no_of_unique=0,
        ratio=0,
    
        # Describe
        self.high=np.nan
        self.low=np.nan
        self.range=np.nan
        
        self.mean=np.nan
        self.median=np.nan
        self.mode=np.nan
        
        self.q1=np.nan
        self.q3=np.nan
        self.iqd=np.nan
        
        self.var=np.nan
        self.std=np.nan
        
        self.skewness=np.nan
        self.kurtosis=np.nan
        
        self.nofbins=np.nan
        
        
    def identify(self):
        self.no_of_unique=len(np.unique(self.data)) 
        self.n=len(self.data)
        self.ratio= float(self.no_of_unique/self.n) *100
        self.ratio=round(self.ratio,2)
        
        self.dtype=str(self.data.dtype)
       
        if "<U" in self.dtype:    
            if self.ratio <=10:
                self.prediction="category"
            try:
                pd.to_datetime(df[col])
                self.prediction="datetime"
            except:
                pass
            
        if "int" in self.dtype:
            if self.ratio <=10:
                self.prediction="int category"
            else:
                self.prediction="feature"
        
        if "float" in self.dtype:
            self.prediction="feature"
            
        
        if self.ratio==100:
            self.prediction="primary key"
    
    
        self.identify_dict={
            "col":self.col,
            "prediction":self.prediction,
            "dtype":self.dtype,
            "n":self.n,
            "no_of_unique":self.no_of_unique,
            "ratio":self.ratio,
        }
        
        return self.identify_dict
        
    def describe(self):
        
        self.identify()
        
        if ("int" in self.dtype) or ("float" in self.dtype):
            self.high=self.data.max()
            self.low=self.data.min()
            self.range=self.high-self.low

            from scipy import stats
            
            self.mean=self.data.mean()
            self.median=np.quantile(self.data, .5)
            self.mode=stats.mode(self.data)[0]
            
            self.q1=np.quantile(self.data, .25)
            self.q3=np.quantile(self.data, .75)
            self.iqd=self.q3-self.q1

            self.var=self.data.var()
            self.std=self.data.std()

            from scipy.stats import kurtosis
            from scipy.stats import skew

            self.skewness=skew(self.data)
            self.kurtosis=kurtosis(self.data)

            self.nofbins= np.round( 1+3.22*np.log10(self.n)+0.5 , 0 )

        self.describe_dict={
            "high":self.high,
            "low":self.low,
            "range":self.range,

            "mean":self.mean,
            "meadian":self.median,
            "mode":self.mode,

            "q1":self.q1,
            "q2":self.q3,
            "q3":self.iqd,

            "var":self.var,
            "std":self.std,

            "skewness":self.skewness,
            "kurtosis":self.kurtosis,

            "nofbins":self.nofbins,
            }

        return self.describe_dict
    
    def charting(self):
#         if ("int" in self.dtype) or ("float" in self.dtype):
#             mychart = Chart(data=self.data, kind='bar')
        pass
        
        
for col in df.columns:
    obj=Colu(df,col,"sample")
    obj.describe()
#     obj.charting()
```
