# Find The Access Codes

```python
def solution(l):
 
  l=sorted(l, reverse=True)
  if len(l) <3:
    return 0
  
  triplet_nos=0
  
  for i in range(len(l)):
    for j in range(i+1,len(l)):
      
      if l[i]%l[j]==0:
        for k in range(j+1,len(l)):
          if l[j]%l[k]==0:
            triplet_nos+=1
  return triplet_nos

l=[1, 2, 3, 4, 5, 6]
solution(l)
```
